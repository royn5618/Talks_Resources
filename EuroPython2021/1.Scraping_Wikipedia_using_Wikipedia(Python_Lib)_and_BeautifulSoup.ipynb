**Caveats**

Tables are unavailable in wikipedia library, so use beautiful soup to parse the html content and extract the tables.

```python
soup = BeautifulSoup(wiki_page_obj.html(), 'html.parser')
```

```python
soup
```

Select the tables using the html table class 'wikiable'

```python
tables = soup.findAll("table", {"class": "wikitable"})
```

Check number of tables in the page:

```python
num_tables = len(tables)
num_tables
```

```python
df = pd.read_html(str(tables))[0]
df.head()
```
